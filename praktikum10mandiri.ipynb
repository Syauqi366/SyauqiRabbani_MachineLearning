```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score


df = pd.read_csv(path + "weather_classification_data.csv")
```

```python
X = df[["Temperature", "Wind Speed"]]
y = df["Cloud Cover"]

le = LabelEncoder()
y_encoded = le.fit_transform(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded, test_size=0.2, random_state=42
)

best_k = 1
best_acc = 0

for k in range(1, 11):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    pred = knn.predict(X_test)
    acc = accuracy_score(y_test, pred)

    if acc > best_acc:
        best_acc = acc
        best_k = k

print("Nilai k terbaik =", best_k)
print("Akurasi terbaik =", best_acc)
```

```python
knn_best = KNeighborsClassifier(n_neighbors=best_k)
knn_best.fit(X, y_encoded)

test_value = pd.DataFrame([[16, 3]], columns=X.columns)
prediction = le.inverse_transform(knn_best.predict(test_value))

print("Persepsi Marry (16°C, 3 km/h):", prediction[0])
```

```python
import pandas as pd
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score
data = {
    "NIM": ["TI001","TI002","TI003","TI004","TI005","TI006","TI007","TI008","TI009","TI010"],
    "Hasil_Sebenarnya": ["Lulus","Lulus","Lulus","Lulus","Lulus",
                         "Tidak Lulus","Tidak Lulus","Tidak Lulus","Tidak Lulus","Tidak Lulus"],
    "Hasil_Prediksi":   ["Lulus","Lulus","Lulus","Tidak Lulus","Tidak Lulus",
                         "Lulus","Tidak Lulus","Tidak Lulus","Tidak Lulus","Tidak Lulus"]
}

df2 = pd.DataFrame(data)

# Konversi label ke bentuk biner
# Lulus = 1, Tidak Lulus = 0
df2['Sebenarnya_bin'] = df2['Hasil_Sebenarnya'].map({"Lulus": 1, "Tidak Lulus": 0})
df2['Prediksi_bin']   = df2['Hasil_Prediksi'].map({"Lulus": 1, "Tidak Lulus": 0})
```

```python
cm = confusion_matrix(df2['Sebenarnya_bin'], df2['Prediksi_bin'])
```

```python
accuracy = accuracy_score(df2['Sebenarnya_bin'], df2['Prediksi_bin'])
precision = precision_score(df2['Sebenarnya_bin'], df2['Prediksi_bin'])
recall = recall_score(df2['Sebenarnya_bin'], df2['Prediksi_bin'])
```

```python
print("=== DATASET ===")
print(df2[['NIM','Hasil_Sebenarnya','Hasil_Prediksi']])
print("\n=== CONFUSION MATRIX ===")
print(cm)
print("\n=== METRIC HASIL ===")
print(f"Accuracy : {accuracy:.2f}")
print(f"Precision: {precision:.2f}")
print(f"Recall   : {recall:.2f}")
```

```python
# Copy dataframe
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler # Import StandardScaler

# Reload the original weather classification data as 'df' was overwritten
weather_df = pd.read_csv(path + "weather_classification_data.csv")
data = weather_df.copy()

# Label Encoding untuk kolom kategorikal
categorical_cols = ["Cloud Cover", "Season", "Location"]
encoder = LabelEncoder()

for col in categorical_cols:
    data[col] = encoder.fit_transform(data[col])

# Pisahkan fitur dan target
X = data.drop("Weather Type", axis=1)
y = data["Weather Type"]

# Encode target
y = LabelEncoder().fit_transform(y)

# Scaling fitur numerik
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_scaled[:5]
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)
```

```python
import numpy as np

errors = []

for k in range(1, 21):
    knn_temp = KNeighborsClassifier(n_neighbors=k)
    knn_temp.fit(X_train, y_train)
    pred_k = knn_temp.predict(X_test)
    errors.append(np.mean(pred_k != y_test))

best_k = np.argmin(errors) + 1
best_k
```

```python
knn = KNeighborsClassifier(n_neighbors=best_k)
knn.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

y_pred = knn.predict(X_test)

print("Akurasi:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

```python
# Contoh data baru (isi sesuai kebutuhan)
sample = pd.DataFrame({
    "Temperature": [28],
    "Humidity": [70],
    "Wind Speed": [5],
    "Precipitation (%)": [20],
    "Cloud Cover": ["clear"],
    "Atmospheric Pressure": [1010],
    "UV Index": [6],
    "Season": ["Summer"],
    "Visibility (km)": [8],
    "Location": ["inland"]
})

# Encode kategorikal
# Re-create and fit LabelEncoder for each categorical column for the sample
# The 'categorical_cols' and 'weather_df' are available from previous cells.
for col in categorical_cols:
    # Concatenate unique categories from original data and the new sample
    all_categories = pd.concat([weather_df[col], sample[col]], axis=0).unique()

    temp_encoder = LabelEncoder()
    temp_encoder.fit(all_categories)
    sample[col] = temp_encoder.transform(sample[col])

# Scaling
sample_scaled = scaler.transform(sample)

# Prediksi
pred_encoded = knn.predict(sample_scaled)[0]

# To inverse transform the prediction, we need the LabelEncoder used for the target variable 'Weather Type'.
# This encoder was not globally stored in previous cells, so we re-create it here.
target_le = LabelEncoder()
target_le.fit(weather_df["Weather Type"]) # Fit on all original Weather Type categories
prediction = target_le.inverse_transform([pred_encoded])[0]

print("Prediksi Weather Type untuk data baru:", prediction)
```

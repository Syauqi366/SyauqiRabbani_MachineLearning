```python
import pandas as pd

df = pd.read_csv(path+f"/data/data.csv")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
print("5 Data Teratas:")
display(df.head())
```

```python
print("\nInformasi Dataset:")
print(df.info())

print("\nStatistika Deskriptif:")
display(df.describe())
```

```python
print("\nNilai unik pada kolom diagnosis:")
print(df['diagnosis'].unique())

print("\nJumlah setiap kelas:")
print(df['diagnosis'].value_counts())
```

```python
df = df.drop(['id', 'Unnamed: 32'], axis=1)

# Label encoding: M = 1 (Malignant), B = 0 (Benign)
le = LabelEncoder()
df['diagnosis'] = le.fit_transform(df['diagnosis'])

# Pisahkan fitur dan label
X = df.drop('diagnosis', axis=1)
y = df['diagnosis']

# Normalisasi fitur agar skala data seragam
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)
```

```python
model = SVC(kernel='linear', random_state=42)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)

print("\nAkurasi Model:")
print(accuracy_score(y_test, y_pred))

print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(5,4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix SVM - Breast Cancer')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x=df['radius_mean'], y=df['texture_mean'], hue=df['diagnosis'], palette='Set1')
plt.title('Visualisasi Dua Fitur: radius_mean vs texture_mean')
plt.show()
```
